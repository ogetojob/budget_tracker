***Expenses Tracker***

```python
items = ["Rent", "Groceries", "Utilities", "Transport", "Entertainment", "Savings"]
amounts = [400, 150, 100, 50, 150, 50]
currency = "USD"
closing_balance = budget - sum(amounts)
categories = ["Housing", "Food", "Utilities", "Transport", "Entertainment"]
is_necessary = [True, True, True, True, False, True]
```

```python
expenses = sum(amounts)
```

```python
len(items)
no_of_items = len(items)
Transactions = no_of_items
```

```python

print("     ***Expenses Report***")
print("=" * 40)
print("Owner".ljust(25), user)
print("Current Month".ljust(25), curent_month)
print("=" * 40)
print("Currency".ljust(25), currency)
print("Opening Balance".ljust(25), str(budget) + " " + currency)
print("=" * 40)

items = {
    "Rent": amounts[0],
    "Groceries": amounts[1],
    "Utilities": amounts[2],
    "Transport": amounts[3],
    "Entertainment": amounts[4],
    "Savings": amounts[5]
}

for k, v in items.items():
    print(k.ljust(25), str(v) + " " + currency)

print("=" * 40)
print("Total Expenses".ljust(25), str(expenses) + " " + currency)
print("Closing Balance".ljust(25), str(closing_balance) + " " + currency)
print("=" * 40)
print("Transaction".ljust(25), Transactions)
print("=" * 40)
print("End of Report")
print("=" * 40)
print("Within_budget".ljust(25), str(is_budget_exceeded).capitalize())
print("=" * 40)
print("Report generated on", timestamp)
```

```python

import matplotlib.pyplot as plt

# Prepare data for plotting
categories = list(items.keys())
amounts = list(items.values())

# Bar chart
plt.bar(categories, amounts)
plt.title("Expenses Breakdown")
plt.xlabel("Category")
plt.ylabel("Amount (USD)")
plt.show()

# Pie chart
plt.pie(amounts, labels=categories, autopct="%1.1f%%", startangle=140)
plt.title("Expenses Distribution")
plt.show()

```
